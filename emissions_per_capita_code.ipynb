```python
import pandas as pd

data = pd.read_csv("./data/ETS_Database_v31.csv", sep="\t")
```

```python
# remove weird date
agg_dates = ['Total 3rd trading period (13-20)', 
             'Total 2nd trading period (08-12)',
             'Total 1st trading period (05-07)']
d1 = data[~data['year'].isin(agg_dates)]
```

```python
pop = pd.read_csv('./data/eu_pop/world_pop.csv')
```

```python
p1 = pop[['Country Name', '2017']]
p1_list = p1['Country Name'].unique()
```

```python
eu_list = d1['country'].unique()[:-1] # remove the strange NER 300 thing
```

```python
eu_list.astype(str) 
```

```python
target_countries = list(set(eu_list) & set(p1_list) ) # removed slovakia lol
```

```python
df_2018 = d1[data['country'].isin(target_countries)]
df_2018 = df_2018[df_2018['year'] == '2017']
df_2018
```

```python
eu_emissions = df_2018.groupby(['country']).sum()
```

```python
# p1 = population of all countries
# eu_emissions = just the emmisions of eu countries in 2018

# Left join
joined = eu_emissions.join(p1.set_index("Country Name"), on="country", how="left")
joined.columns=['emissions', 'population'] # rename cols
```

```python
joined['emissions_per_capita'] = joined['emissions'] / joined['population']
```
